# Complex Mapper Examples

```python
from complex_mapper import *
```

Here is a plot of the Riemann zeta function

```python
plot_format = ComplexGridPlotFormat()
plot_format.set_grid_boundaries([-5.0, 5.0, -np.pi, np.pi])
plot_format.set_plot_boundaries([-12, 12, -6, 6])
plot_format.set_number_of_points_per_line(500)
plot_format.set_domain_grid_spacing(0.05)
```

```python
plot_format.set_title("(z) (Riemann zeta function)")
complex_grid_plot(plot_format, zeta)
```

Here is a plot of the Riemann zeta function along the critical line.

```python
import numpy as np
t = np.linspace(0.0001, 50.0, 1000)
plot_line_in_complex_plane(t, lambda u: 0.5, lambda u: u, zeta, 
                           "Riemann Zeta Function Along Critical Line")
plot_complex_values_along_line(t, lambda u: 0.5*np.ones([len(u)]), lambda u: u, zeta, 
                               "Riemann Zeta Function Along Critical Line")
```

Here are some more functions.

```python
plot_format.set_grid_boundaries([-np.pi/2, np.pi/2, -np.pi/3, np.pi/3])
plot_format.set_plot_boundaries(None)
plot_format.set_number_of_points_per_line(1000)
plot_format.set_domain_grid_spacing(0.05)
plot_format.set_title("z")
complex_grid_plot(plot_format, lambda z: z)
plt.show()
plt.close()
plot_format.set_title("exp(z)")
complex_grid_plot(plot_format, np.exp)
plt.show()
plt.close()
plot_format.set_title("sin(z)")
complex_grid_plot(plot_format, np.sin)
plt.show()
plt.close()
plot_format.set_title("cos(z)")
complex_grid_plot(plot_format, np.cos)
plt.show()
plt.close()
plot_format.set_title("A Polynomial")
complex_grid_plot(plot_format, lambda z: (-10.0 + 10.0j)*z
                                          + (9.0 + 2.0j)*z**2 
                                          + (0.5 + 2.0j)*z**3
                                          - 2*z**4)
plt.show()
plt.close()
```
